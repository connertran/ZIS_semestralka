```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt

# 1) načtení dat
df = pd.read_csv("diabetes_dataset.csv")

print(df)
# 2) převod targetu na 0/1
df["class"] = df["class"].map({"Positive": 1, "Negative": 0})

# 3) převod Yes/No sloupců na 1/0
dataFrameCommon = df.drop(["Age", "Gender", "class"], axis=1)
featuresCommon = dataFrameCommon.columns

for x in featuresCommon:
    df[x] = df[x].map({"Yes": 1, "No": 0})

# 4) převod Gender na čísla
df["Gender"] = df["Gender"].map({"Male": 1, "Female": 0})

# 5) rozdělení na train/test
train, test = train_test_split(df, test_size=0.2, random_state=42, stratify=df["class"])

# 6) rozdělení train dat podle class
# tady jsem musel trošku opravit ty názvy proměnných, protože negative=0 a positive=1, a ne naopak
negative_class = train[train["class"] == 0]
positive_class = train[train["class"] == 1]

# 7) oversampling menší třídy, aby oboje třídy měly stejnou velikost-- přidat další data pro menší třídy
negative_overs = negative_class.sample(
    n=len(positive_class),
    replace=True,
    random_state=42
)

# 8) balanced train dataset
train_balanced = pd.concat([positive_class, negative_overs])

# 9) X a y
x_train = train_balanced.drop(columns=["class"])
y_train = train_balanced["class"]

x_test = test.drop(columns=["class"])
y_test = test["class"]
```

```python
# 10) Random Forest model
# chceme 100 stromů
randomForestClassifier = RandomForestClassifier(
    n_estimators=100,
    max_depth=4,
    random_state=42
)

# 11) trénování
randomForestClassifier.fit(x_train, y_train)

# 12) predikce
pred = randomForestClassifier.predict(x_test)

# 13) evaluace
print("Accuracy:", accuracy_score(y_test, pred))
print("F1 score:", f1_score(y_test, pred))
print("Recall:", recall_score(y_test, pred))
print("Confusion matrix:")
print(confusion_matrix(y_test, pred))
print("Classification report:")
print(classification_report(y_test, pred))
```

```python
from sklearn.tree import plot_tree

# příklad: třeba první strom ze 100 stromů v Random Forest
first_tree = randomForestClassifier.estimators_[0]

plt.figure(figsize=(20, 10))
plot_tree(
    first_tree,
    feature_names=x_train.columns,
    class_names=["Negative", "Positive"],
    filled=True,
    max_depth=4
)
plt.title("Strom č. 0")
plt.show()
```

```python
# příklad: třeba druhý strom ze 100 stromů v Random Forest
second_tree = randomForestClassifier.estimators_[1]

plt.figure(figsize=(20, 10))
plot_tree(
    second_tree,
    feature_names=x_train.columns,
    class_names=["Negative", "Positive"],
    filled=True,
    max_depth=4
)
plt.title("Strom č. 1")
plt.show()
```

```python
#Graf: jak moc každá featura hraje roli v predikci

#random forest umi výpočítat důležitost pro každou featuru z naších 100 stromů

# hodnota důležitosti např. [0.5, 0.32, 0.25, atd...]
importances = randomForestClassifier.feature_importances_
# nazev těch featur
features = x_train.columns

importance_df = pd.DataFrame({
    "feature": features,
    "importance": importances
}).sort_values("importance", ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(importance_df["feature"], importance_df["importance"])
plt.gca().invert_yaxis()
plt.title("Důležitost atributů v Random Forest")
plt.xlabel("Importance")
plt.ylabel("Atribut")
plt.show()

#interpretace výsledku (viz grafu)
  # vidíme, že polydipsia má nejvyšší hodnotu 
    #(např. pokud šlověk má hodně velkouu žízeň, je to vysoká pravděpodobnost, že má diabetu)
```
