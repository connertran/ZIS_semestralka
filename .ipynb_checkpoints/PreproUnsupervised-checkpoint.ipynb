```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score

df = pd.read_csv("diabetes_dataset.csv")

df["class"] = df["class"].map({"Positive": 1, "Negative": 0})

symptom_cols = [col for col in df.columns if col not in ["Age", "Gender", "class"]]
for col in symptom_cols:
    df[col] = df[col].map({"Yes": 1, "No": 0})

df["Gender"] = df["Gender"].map({"Male": 1, "Female": 0})

# podmnožina dát
subset = df[df["Polydipsia"] == 1].copy()
# uloˇzenie pre clustering
subset.to_csv("clustering_dataset.csv", index=False)

# uložiť si class bokom len tak
subset_class = subset["class"]

# clustering vstupy bez targetu
X_cluster = subset.drop(columns=["class"])

# scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# hľadanie vhodného k
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(f"k={k}, silhouette={score:.4f}")

# finálny KMeans
kmeans_final = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans_labels = kmeans_final.fit_predict(X_scaled)

# Agglomerative
agg = AgglomerativeClustering(n_clusters=2)
agg_labels = agg.fit_predict(X_scaled)

print("KMeans silhouette:", silhouette_score(X_scaled, kmeans_labels))
print("Agglomerative silhouette:", silhouette_score(X_scaled, agg_labels))

subset_result = subset.copy()
subset_result["cluster"] = kmeans_labels
subset_result["cluster_agg"] = agg_labels
print(subset_result["cluster"].value_counts())
print("KMeans clusters")
print(subset_result.groupby("cluster").mean(numeric_only=True))

print("Agglomerative clusters")
print(subset_result.groupby("cluster_agg").mean(numeric_only=True))
```
