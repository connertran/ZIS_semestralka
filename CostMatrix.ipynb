```python
import pandas as pd

# confusion matrices zo svojich modelov
cm_tree = [[40, 0],
           [5, 59]]

cm_rf = [[40, 0],
         [4, 60]]

# náklady
cost_TN = 1
cost_FP = 3
cost_FN = 10
cost_TP = 1

def total_cost(cm, tn_cost, fp_cost, fn_cost, tp_cost):
    tn, fp = cm[0]
    fn, tp = cm[1]
    return tn*tn_cost + fp*fp_cost + fn*fn_cost + tp*tp_cost

results = pd.DataFrame([
    {
        "Model": "Decision Tree",
        "TN": cm_tree[0][0],
        "FP": cm_tree[0][1],
        "FN": cm_tree[1][0],
        "TP": cm_tree[1][1],
        "Total cost": total_cost(cm_tree, cost_TN, cost_FP, cost_FN, cost_TP)
    },
    {
        "Model": "Random Forest",
        "TN": cm_rf[0][0],
        "FP": cm_rf[0][1],
        "FN": cm_rf[1][0],
        "TP": cm_rf[1][1],
        "Total cost": total_cost(cm_rf, cost_TN, cost_FP, cost_FN, cost_TP)
    }
])

print(results)
```
